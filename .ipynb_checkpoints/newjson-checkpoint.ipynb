```python
import pandas as pd
import json
from pandas.io.json import json_normalize
```

```python
tp='final_api_data.csv'
all_data=pd.read_csv(tp)
all_data.head()
```

```python
renamed_df = all_data.rename(columns={"tract":"properties.tractce10"})
renamed_df.head()
```

```python
all_data.dtypes
```

```python
cen=list(all_data['tract'])
cen
```

```python
df=pd.read_json('boundaries.geojson')
```

```python
df1=json_normalize(df['features'])
df1
```

```python
df1["properties.tractce10"]=df1["properties.tractce10"].astype(int)
```

```python
df1.dtypes
```

```python
for index, row in df1.iterrows():
    if df1["properties.tractce10"][index] in cen:
        print("okay")
    else:
        df1["properties.tractce10"][index]= "null" 
    
```

```python
df1
```

```python
author = df1["properties.tractce10"].unique()
author
```

```python
edited_df = df1.loc[df1["properties.tractce10"] != "null" , :]
edited_df

```

```python
merge_table = pd.merge(renamed_df, edited_df, on="properties.tractce10")
merge_table.head()
```

```python
merge_table.columns.tolist()
```

```python
organized_df = merge_table[['type', 'properties.tractce10','community_name','poverty_rate', 'home_value', 'home_change', 'total_crimes',  'crimes_per_1000', 'pop_12_17',  'pop_change', 'median_income', 'unemployment', 'assoc_degree_or_higher',  'properties.commarea',  'properties.commarea_n',  'properties.countyfp10',  'properties.geoid10',  'properties.name10', 'properties.namelsad10',  'properties.notes', 'properties.statefp10', 'geometry.type', 'geometry.coordinates' ]]
organized_df.head()
```

```python
organized_df.drop('type', axis=1)
```

```python
organized_df2= organized_df.rename(columns={"geometry.type":"type", "geometry.coordinates":"the_geom", })
organized_df2
```

```python
merge_table.to_json('betterfile.geojson')
organized_df2.to_csv('better2.csv')
```

```python
import pandas as pd
```

```python
tp='OZ-Selected-with-Submissions.csv'
OZ=pd.read_csv(tp)
```

```python
OZ['geoid'] = "14000US" + OZ['DSA Track: Distinct Track Number'].astype(str)
```

```python
ap='pop_census.csv'
pop=pd.read_csv(ap)
```

```python
lp='employment.csv'
employ=pd.read_csv(lp)
```

```python
employ2= employ.drop(['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 5', 'Unnamed: 7', 'Unnamed: 9', 'Unnamed: 11', 'Unnamed: 13', 'Armed Forces', 'In Labor Force'], axis=1)
employ2.head()
```

```python
mp='censustract-00-10.csv'
poppop=pd.read_csv(mp)

 # Using .rename(columns={}) in order to rename columns
    
poppop2 = poppop.rename(columns={"GEOID":"geoid"})
poppop3=poppop2.loc[:,['geoid', 'POP00', 'POP10']]
poppop3.head()
```

```python
pop_df= pd.DataFrame(poppop3)
```

```python
pop_df['geoid']=pop_df['geoid'].astype(str)
```

```python
pop_df['geoid'] = "14000US" + pop_df['geoid'].astype(str)
```

```python
merge_ = pd.merge(pop_df, OZ, on="geoid")
merge_
```

```python
tp='median_census.csv'
income=pd.read_csv(tp)
```

```python
merge_table = pd.merge(pop, merge_, on="geoid")
```

```python
merge2 = pd.merge(merge_table, income, on="geoid")
```

```python
df = merge2.rename(columns={"B01003001":"Population 12-17", "B01003001, Error":"Population, error", "B19013001":"Median Income", "B19013001, Error": "Median Income Error", "name_x": "Census Tract"})
```

```python
df2 = pd.merge(df, employ2, on="geoid")
```

```python
df2['geoid']=df2['geoid'].astype(str)
```

```python
df2= df2.drop(['name'], axis=1)
```

```python
del df2['name_y']
df2.head()
```

```python
aa='df2.csv'
new=pd.read_csv(aa)
new
```

```python
new['Population 12-17']=new['Population 12-17'].astype(int)
new['POP10']=new['POP10'].astype(int)
new['POP00']=new['POP00'].astype(int)
```

```python
percent_change = ((new["Population 12-17"]-new['POP00'])/new['POP00'])*100
new["% Change Pop"] = percent_change
```

```python
new= new.drop(['Unnamed: 0'], axis=1)
```

```python
percent_unem = (new["unemployed"]/new['Civilian Labor Force'])*100
new["% Unemployment"] = percent_unem
```

```python
new.to_csv('new.csv')
new
```

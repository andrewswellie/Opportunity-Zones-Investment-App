```python
import pandas as pd
```

```python
tp='OZcook.csv'
OZ=pd.read_csv(tp)
```

```python
OZ['geoid'] = "14000US" + OZ['Unnamed: 2'].astype(str)
```

```python
ap='pop17C.csv'
pop=pd.read_csv(ap)
pop.head()
```

```python
lp='employmentC.csv'
employ=pd.read_csv(lp)
employ.head()
```

```python
employ2=employ.loc[:, ["geoid", "Civilian Labor Force", "Employed", "Unemployed"]]
employ2.head()
```

```python
mp='censustract-00-10.csv'
poppop=pd.read_csv(mp)

 # Using .rename(columns={}) in order to rename columns
    
poppop2 = poppop.rename(columns={"GEOID":"geoid"})
poppop3=poppop2.loc[:,['geoid', 'POP00', 'POP10']]
poppop3.head()
```

```python
pop_df= pd.DataFrame(poppop3)
```

```python
pop_df['geoid']=pop_df['geoid'].astype(str)
```

```python
pop_df['geoid'] = "14000US" + pop_df['geoid'].astype(str)
```

```python
merge_ = pd.merge(pop_df, OZ, on="geoid")
merge_.head()
```

```python
tp='income.csv'
income=pd.read_csv(tp)
income.head()
```

```python
merge_table = pd.merge(pop, merge_, on="geoid")
```

```python
merge2 = pd.merge(merge_table, income, on="geoid")
```

```python
df = merge2.rename(columns={"B01003001":"Population 12-17", "B01003001, Error":"Population, error", "B19013001":"Median Income", "B19013001, Error": "Median Income Error", "name_x": "Census Tract"})
```

```python
df2 = pd.merge(df, employ2, on="geoid")
```

```python
df2['geoid']=df2['geoid'].astype(str)
```

```python
df2= df2.drop(['name'], axis=1)
```

```python
del df2['name_y']
df2.head()
```

```python
df2['Population 12-17']=df2['Population 12-17'].astype(int)
df2['POP10']=df2['POP10'].astype(int)
df2['POP00']=df2['POP00'].astype(int)
```

```python
percent_change = ((df2["Population 12-17"]-df2['POP00'])/df2['POP00'])*100
df2["% Change Pop"] = percent_change
```

```python
percent_unem = (df2["Unemployed"]/df2['Civilian Labor Force'])*100
df2["% Unemployment"] = percent_unem
```

```python
df2= df2.drop(['Designated Qualified Opportunity Zones'], axis=1)
```

```python
df2.to_csv('new2.csv')
df2
```
